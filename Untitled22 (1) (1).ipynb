```python
# import libraries
from bs4 import BeautifulSoup
import urllib.request
import csv
```

```python
# specify the url
urlpage = 'https://pubmed.ncbi.nlm.nih.gov/?term=intelligence'
```

```python
recherches = []
```

```python
i=1
while i < 10:
    urlpagebis = urlpage + '&page=' + str(i)
    print(urlpagebis)
    page = urllib.request.urlopen(urlpagebis)
    stock = BeautifulSoup(page, 'html.parser')
    recherches.append(stock)
    i+=1
```

```python
enstring = []
```

```python
for a in recherches :
    enstring.append(str(a).split("href="))
```

```python
nbre = []
for i in range(1,10):
    nbre.append(str(i))
nbre
```

```python
b=[]

for i in range (0,len(enstring)):
#     if element[1] == "/":
#         i=1
#         while element[1+i] != "/"
#         i+=1
    
    b.append([element[1:11] for element in enstring[i] if element[1] == "/" and element[2] in nbre])
    
len(b)
```

```python
c = []
for elem in b :
    for elma in elem :
        if elma not in c:
            c.append(elma)
            
len(c)
```

```python
url = 'https://pubmed.ncbi.nlm.nih.gov'
bis = [url + str(element) + '#affiliation-1' for element in c]

len(bis)
```

```python
bis
```

```python
liens = []
for lien in bis :
    # query the website and return the html to the variable 'page'
    page = urllib.request.urlopen(lien)
    # parse the html using beautiful soup and store in variable 'soup'
    soup = BeautifulSoup(page, 'html.parser')
    em_box = soup.find_all({"href":"docsum-title"})
    liens.append(em_box)
```
