```python
auteur_net = []

for i in aut_net:
    auteur_net.append(i.replace("'", "").split(','))

auteur_net
```

```python
auteur_bis = []

for i in range(len(auteur_net)):
    trans = []
    for element in auteur_net[i]:
        if ('\\n') not in element:
            trans.append(element)
    auteur_bis.append(trans)

auteur_bis
```
